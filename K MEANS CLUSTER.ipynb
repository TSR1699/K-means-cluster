```python
from sklearn.cluster import KMeans
```

```python
km1=KMeans()
```

```python
predicted= km1.fit_predict(df[['admin','ms']])
predicted
```

```python
df['initial_clusters']=predicted
```

```python
df.head()
```

```python
centroid= km1.cluster_centers_
```

```python
centroid
```

```python
km1.predict([[15,45]])
```

```python
plt.figure(figsize=(15,10))

plt.scatter(df0['admin'], df0['ms'], color='orange',label='Cluster 0')
plt.scatter(df1['admin'], df1['ms'], color='cyan',label='Cluster 1')
plt.scatter(df2['admin'], df2['ms'], color='pink',label='Cluster 2')
plt.scatter(df3['admin'], df3['ms'], color='green',label='Cluster 3')
plt.scatter(df4['admin'], df4['ms'], color='yellow',label='Cluster 4')
plt.scatter(df5['admin'], df5['ms'], color='black',label='Cluster 5')
plt.scatter(df6['admin'], df6['ms'], color='blue',label='Cluster 6')
plt.scatter(df7['admin'], df7['ms'], color='purple',label='Cluster 7')

#centroids
plt.scatter(centroid[:,0],centroid[:,1], color='red', label="centroid")

plt.legend()
```

```python
wcss=[]
cluster = range(1,15)

for k in cluster:
    km2= KMeans(n_clusters=k)
    km2.fit(df[['admin','ms']])
    wcss.append(km2.inertia_)
wcss
```

```python
plt.plot(cluster, wcss)
```

```python
plt.plot(cluster, wcss,'go--')
```
